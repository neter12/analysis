```python
SQL_Query1 = pd.read_sql_query('''DECLARE @Year1 INT = 2022;

SELECT OPOR.DocDate AS PostDate, ItmsGrpCod, OITM.ItemCode AS ItmCode, ItemName, Quantity
FROM OPOR
JOIN POR1 
ON OPOR.DocEntry = POR1.DocEntry
JOIN OITM
ON POR1.ItemCode = OITM.ItemCode

WHERE YEAR(POR1.DocDate) = @Year1
 ''', con=conn)
```
